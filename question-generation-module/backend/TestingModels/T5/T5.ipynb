```python
from transformers import pipeline

# Load the model
question_generation_model = pipeline("text2text-generation", model="valhalla/t5-small-qg-prepend")
```

```python
# Define a prompt
prompt = "Generate three questions based on this video transcript:\nThis video explains machine learning concepts including supervised and unsupervised learning.\nQuestions:"

# Generate questions
result = question_generation_model(prompt, max_length=200, num_return_sequences=1, num_beams=3)

# Print the generated questions
print("Generated Questions:")
print(result[0]['generated_text'])
```

# valhalla/t5-small-qg-hl

```python
from transformers import pipeline

# Load the model fine-tuned for question generation
question_generation_model = pipeline("text2text-generation", model="valhalla/t5-small-qg-hl")
```

```python
# Define text with <hl> tags
text = (
    "The video explains <hl>supervised learning<hl>. "
    "Supervised learning involves labeled data to train a model, and examples include regression and classification."
)

# Generate questions
result = question_generation_model(text, max_length=128, num_beams=5)
print(result[0]['generated_text'])
```
